```python
X = Hour.drop(columns = ['cnt','windspeed','weathersit','workingday','weekday','holiday','season'],axis=1) #feature
Y = Hour['cnt'] #label
len(X.columns)
```

```python
# splitting data into train and test 
from sklearn.model_selection import train_test_split
X_train, x_test,Y_train,y_test = train_test_split(X,Y,test_size=0.2 , random_state=0) #splitted data for validation
```

```python
# scaling the values 
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
x_test = scale.transform(x_test)
```

```python
# building model \
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
LR = LinearRegression().fit(X_train,Y_train)
y_pred = LR.predict(x_test)
cv = cross_val_score(LR, X_train,Y_train, cv=10)
cv.mean()
```

```python
LR.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
DR = DecisionTreeRegressor()
DR.fit(X_train,Y_train)
```

```python
y_predDR = DR.predict(x_test)
```

```python
DR.score(X_train,Y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
MR = RandomForestRegressor(n_estimators=100).fit(X_train,Y_train)
y_predmr = MR.predict(x_test)
MR.score(X_train,Y_train)
```

```python
print(f"accuracy of model Random :{MR.score(x_test,y_test)}")
```

```python
# calculating RMSE (Root mean squared error)
from sklearn.metrics import mean_squared_error
import numpy as np
mse = mean_squared_error(y_test,y_predmr)
rmse = np.sqrt(mse)
print(f"rmse of model Random : {rmse}")
```
